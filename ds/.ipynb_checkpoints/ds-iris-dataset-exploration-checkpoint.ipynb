## Maßzahlen

Als ersten Hinweis auf die Güte eines Datensatzes sind einfache Maßzahlen oft ein hilfreiches Mittel. Pandas bietet einfache Möglichkeiten um Mittelwerte, Schwankungsbreiten und andere Zahlen zu ermitteln.

````python
def analyse_column(name):
    col = df[name]
    
    print(
        'Die Verteilung der Spalte {}\n'.format(name) + 
        'hat {} Einträge\n'.format(col.count()) +
        'und einen Mittelwert von {:.2f}\n'.format(col.mean()) +
        'sowie eine Breite von {:.2f}.\n'.format(col.std()) +
        'Das entspricht einem statistischen Fehler von {:.2f}.\n###\n'.format(col.std()/np.sqrt(col.count())) 
    )
    
analyse_column('sepal length (cm)')
analyse_column('petal length (cm)')
````

Hier wurden folgende Kenngrößen berechnet:
  * Stichprobengröße ```col.count()```
  * Arithmetischer Mittelwert ```col.mean()```
  * Breite der Verteilung ```col.std()```
  * Fehler auf den Mittelwert ```col.std()/np.sqrt(col.count()```

## Erweiterung: Trennung nach Klassen

Wir erweitern die Funktion ```analyse_column``` um die Möglichkeit, die Maßzahlen nur für einzelne Arten zu berechnen.

````python
def analyse_column_kind(name, kind=None):
    if kind is None:
        col = df[name]
    else:
        col = df[df['target'] == kind][name]
    
    print(
        'Die Verteilung der Spalte {}, Typ {}\n'.format(name,kind) + 
        'hat {} Einträge\n'.format(col.count()) +
        'und einen Mittelwert von {:.2f}\n'.format(col.mean()) +
        'sowie eine Breite von {:.2f}.\n'.format(col.std()) +
        'Das entspricht einem statistischen Fehler von {:.2f}.\n###\n'.format(col.std()/np.sqrt(col.count())) 
    )
    
analyse_column_kind('sepal length (cm)',0)
analyse_column_kind('sepal length (cm)',1)
analyse_column_kind('sepal length (cm)',2)
````

### Diskussion

Die Unterschiedlichen Arten unterscheiden sich in den zentralen Parametern. Dies ermöglicht eine Klassifizierung. Durch die hohen Breiten der Verteilungen überlappen die Arten in den gemessenen Parametern. Diese Überlappung kann durch Histogramme sichtbar gemacht werden.

## Erste Plots

````python
variable = 'sepal length (cm)'

ax=plt.subplots(figsize=(12,8))

for target in df['target'].unique():
    #plt.hist() erzeugt ein Histogramm über eine Liste von Werten
    plt.hist(
        df[df['target'] == target][variable],
        alpha=0.5, bins=np.linspace(3,9,13)
    )
    plt.xlabel(variable)
````

### Neues Python Element: ```plt.hist()```

Mit ```plt.hist(y)``` wird ein Histogram über die Werte in der Liste ```y``` erzeugt. Die Histogramm-Funktion hat verschiedene nützliche Parameter:

  * ```bins```: Der Parameter ```bins=np.linspace(4,8,9)``` erzeugt 9 Bingrenzen die gleichmäßig über das Intervall 4-8 verteilt sind. Also 8 Bins mit einer jeweiligen Breite von 0.5. Die funktion ```np.linspace()``` ist eine nützliche Funktion aus ```numpy``` (```np```).
  * ```alpha```: Der Alpha-Wert zwischen 0 und 1 gibt die Transparenz der Darstellung des jeweiligen Plots an (0=transparent, 1=nicht-transparent). Damit übereinanderliegende Elemente trotzdem sichtbar bleiben wählen wir hier ```alpha=0.5```

## Trennkraft zwischen jeweils zwei Variablen

### (Python: Die gridspec-Bibliothek zum Kombinieren von Plots)

Der Iris Datensatz beinhaltet vier verschiedene Variablen: Die Längen und Breiten von Blüten- und Kelchblättern. In diesem Beispiel kombinieren wir jeweils zwei der Variablen in einer Darstellung.

````python
#Eine Grafik anlegen
plt.figure(figsize=(12,8))

#Ein 2x2 Grid für 4 Plots aufbauen
#Die Plots sollen sich die Achsen teilen, und jeweils gleiche Höhen und Breiten haben.
gs1 = gridspec.GridSpec(2, 2, height_ratios=[1, 1], width_ratios=[1, 1])
gs1.update(wspace=0.0, hspace=0.0)

#Das erste der vier Grid-Felder ansteuern
ax1 = plt.subplot(gs1[0])
#Den Plot erstellen und auf das Feld zeichnen.
plt.scatter(df['sepal width (cm)'],df['sepal length (cm)'],
            c=df['target'],cmap=plt.cm.Set1,s=50)

#...
ax2 = plt.subplot(gs1[1])
plt.scatter(df['petal width (cm)'],df['sepal length (cm)'],c=df['target'],cmap=plt.cm.Set1,s=50)
ax3 = plt.subplot(gs1[2])
plt.scatter(df['sepal width (cm)'],df['petal length (cm)'],c=df['target'],cmap=plt.cm.Set1,s=50)
ax4 = plt.subplot(gs1[3])
plt.scatter(df['petal width (cm)'],df['petal length (cm)'],c=df['target'],cmap=plt.cm.Set1,s=50)

#Alle Achsenbeschriftungen die nicht sichtbar sein sollen verstecken.
###
### Die folgenden Zeilen auskommentieren um den Effekt zu sehen
###
setp( ax1.get_xticklabels(), visible=False)
setp( ax2.get_xticklabels(), visible=False)
setp( ax2.get_yticklabels(), visible=False)
setp( ax4.get_yticklabels(), visible=False)
###

#Die Achsentitel hinzufügen, so das jede sichtbare Achse beschriftet ist
#Diese Achsen sind genau die vier die nicht ausgeblendet werden
ax1.set_ylabel('sepal length (cm)')
ax3.set_ylabel('petal length (cm)')
ax3.set_xlabel('sepal width (cm)')
ax4.set_xlabel('petal width (cm)')

#Magischer Befehl der "meistens" zu besseren Plotergebnissen führt.
plt.tight_layout()
````

### Diskussion des Plots

Je nachdem welche Variablen kombiniert werden, liegen die Datensätze weiter auseinander oder näher beisammen. Für ML Ansätze bedeutet das: Je mehr Trennung in verschiedenen Kombinationen von Eingangsgrößen erkennbar ist, desto besser kann ein ML-Algorithmus klassifizieren.
Wir suchen nach der Datenaufbereitung oft Variablen die ein gewisses Maß an Trennung aufweisen.

## Die Bedeutung von Korellationen

Einfache Maßzahlen verlieren an Bedeutung, wenn Datensätze ungewöhnliche Eigenschaften aufweisen oder wenn Einflussgrößen nicht unabhängig voneinander sind. Korellationen machen vieles komplizierter, liefern aber auch eine große Menge an Informationen.

````python
#Bibliothek zum Zeichnen von Ellipsen
from matplotlib.patches import Ellipse

#Vereinfachung: Daten für x und y Datensätze
x=df[df['target'] == 1]['sepal width (cm)']
y=df[df['target'] == 1]['petal width (cm)']

#Eine Grafik anlegen
plt.figure(figsize=(8,8))

#Ein 2x2 Grid für 4 Plots aufbauen
#Die Plots sollen sich die Achsen teilen, und jeweils gleiche Höhen und Breiten haben.
gs1 = gridspec.GridSpec(2, 2, height_ratios=[1, 3], width_ratios=[3, 1])
gs1.update(wspace=0.0, hspace=0.0)

#Die Einzelteile der Darstellung plotten

#Der Scatter Plot
ax2 = plt.subplot(gs1[2])
plt.scatter(x,y,s=50)
plt.xlabel('sepal width (cm)')
plt.ylabel('petal width (cm)')
#Das obere Histogramm
ax0 = plt.subplot(gs1[0], sharex=ax2)
plt.hist(x)
setp( ax0.get_xticklabels(), visible=False)
#Das rechte Histogramm
ax3 = plt.subplot(gs1[3], sharey=ax2)
plt.hist(y, orientation='horizontal')
setp( ax3.get_yticklabels(), visible=False)

#Berechnung der Parameter für Korellationsellipsen
eigenvectors = np.linalg.eig(np.cov(x,y))
ev = eigenvectors[0]
ex = eigenvectors[1][0]
ey = eigenvectors[1][1]

angle=arctan(ey[0]/ey[1])/np.pi*180

#Ellipsen für 1 & 2 sigma Intervalle zeichnen
ax2.add_artist(Ellipse(
        xy=(x.mean(),y.mean()),
        width=np.sqrt(6 * ev[0]),
        height=np.sqrt(6 * ev[1]),
        angle=angle, alpha=0.30))
ax2.add_artist(Ellipse(
        xy=(x.mean(),y.mean()),
        width=2*np.sqrt(6 * ev[0]),
        height=2*np.sqrt(6 * ev[1]),
        angle=angle, alpha=0.30))
````

### Diskussion des Plots

Der Plot zeigt eine Auswahl von zwei Variablen des Datensatzes und die Korellation zwischen beiden. Die hellblauen und dunkelblauen Ellipsen zeigen die ein (zwei) sigma Umgebung des Mittelwerts der Veteilung. Diese soll 68% (95%) der angezeigten Daten enthalten.

Die Histogramme zeigen eine Projektion der zweidimensionalen Verteilung auf die jeweilige Variable.
